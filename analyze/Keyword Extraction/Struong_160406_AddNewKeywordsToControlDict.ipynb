```python
__author__ = 'a_medelyan'

# Given a flat list of terms in a text file, creates an SKOS rdf file


def write_skos_header(output_file, header_file, url):
    with open(header_file) as f:
        with open(output_file, "w") as f1:
            for line in f:
                if 'URL' in line:
                    line = line.replace('URL', url)
                f1.write(line)
            f1.write('\n')


def convert_text_to_skos(text_file, skos_file, url):
    output_file = open(skos_file, 'a')
    with open(text_file, 'Ubr') as f:
        count = 0
        for line in f:
            term = line.rstrip().decode('latin-1').encode('utf-8')

            if '&' in term:
                term = term.replace('&', '&amp;')

            count += 1

            output_file.write('\t<skos:Concept rdf:about=\"' + url + '/' + str(count) + '\">\n')
            output_file.write('\t\t<skos:prefLabel xml:lang="en">' + term + '</skos:prefLabel>\n')

            output_file.write('\t</skos:Concept>\n')

    f.close()

    output_file.write('</rdf:RDF>\n')
    output_file.close()

```

```python
def convert_dict_to_skos(dico, skos_file, url):
    output_file = open(skos_file, 'wb')
    
    count = 10000
    for cat,kwlist in dico.iteritems():
        count+=1
        output_file.write('\t<skos:Concept rdf:about=\"' + url + '/' + str(count) + '\">\n')
        output_file.write('\t\t<skos:prefLabel xml:lang="en">' + cat + '</skos:prefLabel>\n')
        count2=count
        for kw in kwlist:
            count2+=1
            output_file.write('\t\t<skos:narrower rdf:resource=\"'+ url + '/' +str(count2)+'\"/>\n')
        output_file.write('\t</skos:Concept>\n')
        
        count3=count
        for kw in kwlist:
            count += 1
            output_file.write('\t<skos:Concept rdf:about=\"' + url + '/' + str(count) + '\">\n')
            output_file.write('\t\t<skos:prefLabel xml:lang="en">' + kw + '</skos:prefLabel>\n')
            output_file.write('\t\t<skos:broader rdf:resource=\"'+ url + '/' + str(count3) +'\"/>\n')
            output_file.write('\t</skos:Concept>\n')



    output_file.write('</rdf:RDF>\n')
    output_file.close()
```
